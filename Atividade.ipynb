```python
import json
from datetime import datetime

# Lista de tarefas
tarefas = []

def adicionarTarefa(tarefa, prioridade, data_vencimento):
    """Adiciona uma nova tarefa à lista de tarefas."""
    tarefas.append({
        'tarefa': tarefa,
        'concluida': False,
        'prioridade': prioridade,
        'data_vencimento': data_vencimento
    })
    print(f"Tarefa '{tarefa}' adicionada com sucesso!")

def removerTarefa(tarefa):
    """Remove uma tarefa da lista de tarefas."""
    for t in tarefas:
        if t['tarefa'] == tarefa:
            tarefas.remove(t)
            print(f"Tarefa '{tarefa}' removida com sucesso!")
            return
    print(f"Tarefa '{tarefa}' não encontrada na lista!")

def concluirTarefa(tarefa):
    """Marca uma tarefa como concluída."""
    for t in tarefas:
        if t['tarefa'] == tarefa:
            t['concluida'] = True
            print(f"Tarefa '{tarefa}' marcada como concluída!")
            return
    print(f"Tarefa '{tarefa}' não encontrada na lista!")

def listarTarefa():
    """Exibe todas as tarefas na lista."""
    if not tarefas:
        print("Nenhuma tarefa na lista.")
        return
    for t in tarefas:
        status = "Concluída" if t['concluida'] else "Não concluída"
        print(f"Tarefa: {t['tarefa']} - Status: {status} - Prioridade: {t['prioridade']} - Data de Vencimento: {t['data_vencimento']}")

def listarTarefaPorStatus(concluida):
    """Lista tarefas com base em seu status de conclusão."""
    status_str = "concluídas" if concluida else "não concluídas"
    print(f"Tarefas {status_str}:")
    tarefas_filtradas = [t for t in tarefas if t['concluida'] == concluida]
    if not tarefas_filtradas:
        print(f"Nenhuma tarefa {status_str}.")
    for t in tarefas_filtradas:
        print(f"Tarefa: {t['tarefa']} - Prioridade: {t['prioridade']} - Data de Vencimento: {t['data_vencimento']}")

def salvarTarefas():
    """Salva a lista de tarefas em um arquivo JSON."""
    with open('tarefas.json', 'w') as file:
        json.dump(tarefas, file)
    print("Tarefas salvas com sucesso!")

def carregarTarefas():
    """Carrega a lista de tarefas de um arquivo JSON."""
    try:
        with open('tarefas.json', 'r') as file:
            global tarefas
            tarefas = json.load(file)
        print("Tarefas carregadas com sucesso!")
    except FileNotFoundError:
        print("Nenhum arquivo de tarefas encontrado.")

def menu():
    """Exibe o menu interativo para o usuário."""
    carregarTarefas()  # Carrega as tarefas ao iniciar o programa
    while True:
        print("Menu de Gerenciamento de Tarefas")
        print("1. Adicionar Tarefa")
        print("2. Remover Tarefa")
        print("3. Concluir Tarefa")
        print("4. Listar Todas as Tarefas")
        print("5. Listar Tarefas Concluídas")
        print("6. Listar Tarefas Não Concluídas")
        print("7. Salvar Tarefas")
        print("8. Sair")
        
        escolha = input("Escolha uma opção: ")

        if escolha == '1':
            tarefa = input("Digite a descrição da tarefa: ")
            prioridade = input("Digite a prioridade da tarefa (alta, média, baixa): ")
            data_vencimento = input("Digite a data de vencimento da tarefa (dd/mm/aaaa): ")
            adicionarTarefa(tarefa, prioridade, data_vencimento)
        elif escolha == '2':
            tarefa = input("Digite a descrição da tarefa a ser removida: ")
            removerTarefa(tarefa)
        elif escolha == '3':
            tarefa = input("Digite a descrição da tarefa a ser concluída: ")
            concluirTarefa(tarefa)
        elif escolha == '4':
            listarTarefa()
        elif escolha == '5':
            listarTarefaPorStatus(True)
        elif escolha == '6':
            listarTarefaPorStatus(False)
        elif escolha == '7':
            salvarTarefas()
        elif escolha == '8':
            print("Saindo do programa...")
            salvarTarefas()  # Salva as tarefas ao sair do programa
            break
        else:
            print("Opção inválida! Por favor, escolha uma opção válida.")

# Executa o menu interativo
menu()
```
